```python
# Dependencies and Setup
import pandas as pd


# Read Purchasing File and store into Pandas data frame
file = "04-Pandas_Homework_Instructions_HeroesOfPymoli_Resources_purchase_data.csv"
data = pd.read_csv(file)
data.dropna(inplace = True)
```

## Player Count

```python
#find the total players w/o repeating
total_players = len(data['SN'].value_counts())

#print out
Total_Players_df = pd.Series(['Total Players',total_players])
Total_Players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#count 'item id' w/o repeating
unique_items = len(data['Item ID'].value_counts())
#find average price
avg_price = data['Price'].mean()
#number of purchases
number_purchased = data['Purchase ID'].count()
#revenue = sum(price)
revenue = data['Price'].sum()

#create table and formatting
purchasing_analysis_total = pd.DataFrame({'Number of Unique Item':[unique_items],'Average Price':[f'${round(avg_price,2)}'],'Number of Purchases':[number_purchased],'Total Revenue':[f'${round(revenue,2)}']})

purchasing_analysis_total
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#only male
male_data = data.loc[data['Gender'] == 'Male', :]
male = len(male_data['SN'].value_counts())

#only female
female_data = data.loc[data['Gender'] == 'Female', :]
female = len(female_data['SN'].value_counts())

#other 
other_data = data.loc[data['Gender'] == 'Other / Non-Disclosed', :]
other = len(other_data['SN'].value_counts())

#define percentage(a,b)
def percentage(a, b):
    return format(a / b * 100, '.2f')

#find percentage for each category and formatting
percentage_male = f'{percentage(male,576)}%'
percentage_female = f'{percentage(female,576)}%'
percentage_other = f'{percentage(other,576)}%'

#create table
gender_demographics = pd.DataFrame({'Gender':['Male','Female','Other / Non-Disclosed'],'Total Count':[male,female,other],'Percentage of Players':[percentage_male,percentage_female,percentage_other]})

gender_demographics

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#purchase count for each category
male_p_count = male_data['Purchase ID'].count()
female_p_count = female_data['Purchase ID'].count()
other_p_count = other_data['Purchase ID'].count()

#average purchase price for each category and formatting
male_avg_price = f'${round(male_data["Price"].mean(),2)}'
female_avg_price = f'${round(female_data["Price"].mean(),2)}'
other_avg_price = f'${round(other_data["Price"].mean(),2)}'

#total purchase for each category and formatting
male_total_p = f'${male_data["Price"].sum()}'
female_total_p = f'${female_data["Price"].sum()}'
other_total_p = f'${other_data["Price"].sum()}'

#Avg Total Purchase per Person for each category = total purchase / number of people; and formatting
male_avg_p = f'${round(male_data["Price"].sum()/male,2)}'
female_avg_p = f'${round(female_data["Price"].sum()/female,2)}'
other_avg_p = f'${round(other_data["Price"].sum()/other,2)}'

#create table
purchasing_analysis_gender = pd.DataFrame({'Gender':['Male','Female','Other / Non-Disclosed'],
                                           'Purchase Count':[male_p_count,female_p_count,other_p_count],
                                           'Average Purchase Price':[male_avg_price,female_avg_price,other_avg_price],
                                           'Total Purchase Value':[male_total_p,female_total_p,other_total_p],
                                           'Avg Total Purchase per Person':[male_avg_p,female_avg_p,other_avg_p]})
purchasing_analysis_gender
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#groupby 'SN'
spenders_df = data.groupby(['SN'])

#define the 3 variables
purchase_count = spenders_df['Price'].count()
total_purchase_value = round(spenders_df["Price"].sum(),2)
avg_purchase_price = round(total_purchase_value / purchase_count,2)

#build a table and add columns into it 
top_spender_df = pd.DataFrame(total_purchase_value)
top_spender_df['Purchase Count'] = purchase_count
top_spender_df['Average Purchase Price'] = avg_purchase_price
top_spender_df.columns = ['Total Purchase Value','Purchase Count','Average Purchase Price']

#find the 5 top spenders regarding 'total purchase value'
top_spender = top_spender_df.nlargest(5, ['Total Purchase Value']) 

#format the table
top_spender['Total Purchase Value'] = top_spender['Total Purchase Value'].map('${}'.format)
top_spender['Average Purchase Price'] = top_spender['Average Purchase Price'].map('${}'.format)
column_names = ['Purchase Count','Average Purchase Price','Total Purchase Value']
top_spender.reindex(columns=column_names) #sorry I have a question here:why it did not rearrange the columns here??

top_spender
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Retrieve the Item ID, Item Name, and Item Price columns
popular_df = data.loc[:, 'Item ID':'Price'].copy()
#Group by Item ID and Item Name
popular_df = popular_df.groupby(['Item ID','Item Name'])
popular_df.head(30)
#build table based on price
popular_1_df = data[['Item ID', 'Item Name', 'Price']]
popular_1_df = popular_1_df.drop_duplicates(['Item ID'])
popular_1_df.set_index(['Item ID', 'Item Name'], inplace=True)
popular_1_df.head(30)

#define variables
purchase_count = round(popular_df['Item ID'].count(),2)
total_purchase_value = round(popular_df['Price'].sum(),2)

#build a table and add columns into it 
popular_2_df = pd.DataFrame(purchase_count)
popular_2_df['Total Purchse Value'] = total_purchase_value
popular_2_df.columns = ['Purchase Count', "Total Purchase Value"]
# combine 2 tables
popular_3_df = popular_1_df.join(popular_2_df)


#find the 5 most popular items regarding 'purchase count'
popular = popular_3_df.nlargest(5, ['Purchase Count']) 

#format the table
popular['Total Purchase Value'] = popular['Total Purchase Value'].map('${}'.format)
popular['Price'] = popular['Price'].map('${}'.format)
column_names = ['Purchase Count','Price','Total Purchase Value']
popular.reindex(columns=column_names) #same question here:why it did not rearrange the columns here??

popular
```
